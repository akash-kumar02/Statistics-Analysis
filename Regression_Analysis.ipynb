```python
import pandas as pd
import numpy as np
```

```python
data={
    "Age":[25,30,35,40,45,50],
    "Work exp":[2,5,8,12,16,20]
}
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
sum_xy=(df["Age"]*df["Work exp"]).sum()
sum_xy
```

```python
sum_x=df["Age"].sum()
sum_x
```

```python
sum_y=df["Work exp"].sum()
sum_y
```

```python
s=sum_x*sum_y
s
```

```python
x_squre=(df["Age"]**2).sum()
x_squre
```

```python
squre=(sum_x)**2
squre
```

```python
n=len(df)
n
```

```python
a=n*sum_xy
a
```

```python
b=n*x_squre
b
```

```python
slope=round((a-s)/(b-squre),2)
slope
```

```python
intercept=(sum_y-(slope*sum_x))/n
intercept
```

```python
# Regression equation
y=slope*df["Age"]+intercept
y
```

```python
y=round(slope*38+intercept,2)
y
```
