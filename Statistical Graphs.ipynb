**Bar Graphs**

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
categories=['A','B','C','D','E']
values=[5,7,9,3,6]
plt.bar(categories,values,color='yellow',edgecolor='black')
plt.xlabel("Students Name")
plt.ylabel("Marks in Stats")
plt.title("Example bar chart")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,7,9,3,6]
widths=[1,1.5,1.2,1.1,1.3]
colors=['yellow','lime','skyblue','gold','hotpink']
plt.bar(categories,values,width=widths,color=colors,edgecolor='black')
plt.xlabel("Students Name")
plt.ylabel("Marks in Stats")
plt.title("Example bar chart")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,7,9,3,6]
widths=[1,1.5,1.2,1.1,1.3]
colors=['yellow','lime','skyblue','gold','hotpink']
plt.barh(categories,values,color=colors,edgecolor='black')
plt.xlabel("Students Name")
plt.ylabel("Marks in Stats")
plt.title("Example bar chart")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,7,9,3,6]
plt.bar(categories,values,color='yellow',edgecolor='black')
plt.xlabel("Students Name")
plt.ylabel("Marks in Stats")
plt.title("Example bar chart")
plt.grid(axis='y',linestyle='--')
plt.show()
```

Line **Graph**

```python
x=[2,5,6,10,8,15,7]
y=[4,7,3,9,7,12,13]
plt.plot(x,y,color='blue',linestyle='--',marker='*',label='data points')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example line chart")
```

```python
x=[2,5,6,10,8,15,7]
y=[4,7,3,9,7,12,13]
plt.plot(x,y,color='blue',linestyle='--',marker='*',label='data points')
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example line chart")
```

```python
x=[2,4,5,7,9,12,15]
y=[4,7,8,9,11,12,13]
plt.scatter(x,y,color='blue',linestyle='--',marker='*',label='y=x*2')
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example scatter chart")
```

```python
x=[2,4,5,7,9,12,15]
y=[4,7,8,9,11,12,13]
plt.scatter(x,y,color='blue',linestyle='--',alpha=0.7,marker='*',edgecolors='teal',label='data points')
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example scatter chart")
```

**Histogram**

```python
import matplotlib.pyplot as plt
data=np.random.randn(1000)
plt.hist(data,bins=20,color='b',alpha=0.7,edgecolor='black')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('basic histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
np.random.seed(30)
data=np.random.randn(1000)
plt.hist(data,bins=20,color='lime',alpha=0.7,edgecolor='black',label='random number')
plt.legend()
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('basic histogram')
plt.show()
```

```python
# plot pie chart
sizes=[40,30,20,10]
labels=['Group 1','Group 2','Group 3','Group 4']
colors=['skyblue','orange','green','red']
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140,colors=colors)
plt.title('pie chart example')
plt.show()
```

```python
sizes=[40,30,20,10]
labels=['Group 1','Group 2','Group 3','Group 4']
colors=['skyblue','orange','green','red']
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140,colors=colors,explode=(0,0.1,0,0.23))
plt.title('pie chart example')
plt.show()
```

```python
w=0.2
x=['C1','C2','C3','C4']
boys=[100,50,80,60]
girls=[150,90,50,70]
bar1=np.arange(len(x))
plt.bar(bar1,boys,w,label='boys')
bar2=[i+w for i in bar1]
plt.bar(bar2,girls,w,label='girls')
plt.xticks(bar1+w/2,x)
plt.show
```
