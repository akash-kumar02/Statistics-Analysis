```python
from scipy.stats import shapiro

# Example Dataset
data=[12,43,20.5,56,54,14.6,2,67,5.6]

# perform the shapiro-test
stat,p=shapiro(data)

# print the result

print("shapiro-wilk test statistic:",stat)
print("p-value:",p)
if p>0.05:
  print("the data is normally distributed(fail to reject Ho)")
else:
  print("the data is not normally distributed(reject Ho)")
```

```python
from scipy.stats import normaltest
import numpy as np

# Example Dataset
data=[12,43,20.5,56,54,14.6,2,67,5.6]

# perform the  D'Asotino and pearson k**2 test
stat,p = normaltest(data)

# print the result

print("k**2  statistic:",stat)
print("p-value:",p)
if p>0.05:
  print("the data is normally distributed(fail to reject Ho)")
else:
  print("the data is not normally distributed(reject Ho)")
```
